```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {
    "Sales": [10,20,30,40],
    "Profits":[2,4,3,1],
    "Date":["2022-01-01", "2022-02-01", "2022-03-01", "2020-04-01"]
}

df = pd.DataFrame(data=data_dict, index=data_dict["Date"], columns=["Sales", "Profits"])
```

```python
df
```

```python
fig, ax = plt.subplots();

ax.plot(df);
ax.set_title("Product Sales and profits");
```

```python
# 2nd method to do the same

fig, ax = plt.subplots();

ax.plot(df.index, df["Sales"]);
ax.plot(df.index, df["Profits"], ls="--");
ax.set_title("Product Sales and profits");
```
