# Coffee Production stats
Create a bar chart of the top 10 coffee producers by total production in 'total-production.csv', sorted from highest to lowest. Change the xticks to two letter country abbreviations 

```python
coffee_production = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/total-production.csv").T
coffee_production.columns = coffee_production.iloc[0]
coffee_production = coffee_production.drop("total_production")
```

# Composition over time
Our client is interested in looking at how Brazil's share of global coffee production has changed over time.

Build a plot that global total production as well as Brazil's contribution to global production over time.

```python
brazil_vs_others = (coffee_production.assign(rest_of_world = coffee_production.drop("Brazil", axis=1).sum(axis=1))\
                                        .loc[:,["Brazil", "rest_of_world"]].astype({"Brazil":"float64"}))
brazil_vs_others.dtypes
```

```python
brazil_vs_others.head()
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    brazil_vs_others.index.astype("datetime64[ns]"),
    brazil_vs_others["Brazil"],
    brazil_vs_others["rest_of_world"].astype("float64"),
    labels=["Brazil", "World Total"]
)

fig.suptitle("Brazil's shares of Global coffee production 1990-2018")
ax.set_title("Brazil's increase Share of Global Coffee Market")
ax.set_ylabel("Production (60Kg Bags)", fontsize=14)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.legend(loc="upper left");
```

# Composition of the coffee production

Create a pie chart with top 5 producers, as well as the sum of the rest of the countries.

Finally, create two donut charts, one representing Brazil's share of global production in 1990, and one for brazil's share in 2018.

```python
coffee_production["rest_of_world"] = coffee_production.drop(["Brazil", "Viet Nam", "Colombia", "Indonesia", "Ethiopia"]
                                                            , axis = 1).sum(axis=1)

```

```python
coffee_production_top5_2018 = coffee_production.loc['2018',["Brazil", "Viet Nam", "Colombia", "Indonesia",
                                                            "Ethiopia", "rest_of_world"]]
```

```python
fig, ax = plt.subplots()

ax.pie(
    x = coffee_production_top5_2018[::-1],
    startangle=90,
    labels=coffee_production_top5_2018.index[::-1],
    autopct = "%.0f%%"
);

fig.suptitle("Share of Global Coffee Production 2018");
```

```python
brazil_vs_others = (coffee_production.assign(rest_of_world = coffee_production.drop("Brazil", axis=1).sum(axis=1))\
                                        .loc[:,["Brazil", "rest_of_world"]].astype({"Brazil":"float64"}))
brazil_vs_others.head()
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[0].sort_values(ascending=False),
    startangle=90,
    labels=["",""],
#     autopct="%.0f%%",
    pctdistance=.85,
    colors = ["white", "blue"]
);

hole = plt.Circle((0,0), 0.70, fc = 'white')
fig = plt.gcf()

fig.gca().add_artist(hole)

plt.text(
    0,
    0,
    f"{round((brazil_vs_others.loc['1990', 'Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%",
    ha="center",
    va="center",
    fontsize = 30
)

ax.set_title("Brazil Share Of Global Production 1990");
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[-1].sort_values(ascending=False),
    startangle=90,
    labels=["",""],
#     autopct="%.0f%%",
    pctdistance=.85,
    colors = ["white", "blue"]
);

hole = plt.Circle((0,0), 0.70, fc = 'white')
fig = plt.gcf()

fig.gca().add_artist(hole)

plt.text(
    0,
    0,
    f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%",
    ha="center",
    va="center",
    fontsize = 30
)

ax.set_title("Brazil Share Of Global Production 2018");
```

# Coffee Consumption Vs Prices

Compare Coffee prices with coffee consumption in the following countries.

Is there any relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price Vs consumption totals for these countries.

```python
imports = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/imports.csv")
```

```python
imports.head()
```

```python
consumption = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/imports.csv").set_index("imports").mean(axis=1)

consumption.name = "imports"

consumption.head()
```

```python
prices = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/retail-prices.csv")

prices = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/retail-prices.csv").set_index("retail_prices").mean(axis=1)

prices.name = "prices"

prices.head()
```

```python
price_cons = pd.merge(consumption, prices, how='inner', left_on=consumption.index, right_on=prices.index)\
                .sort_values(by="imports", ascending=False)

price_cons.columns = ["country", "Imports", "Price"]

price_cons.head()
```

```python
fig, ax = plt.subplots(figsize = (14, 10))

width = .35
x = np.arange(0, len(price_cons["country"]))


bar1 = ax.bar(
    x-width/2,
    price_cons["Imports"],
    width = width
)

ax.set_title("Prices Paid Vs. Consumption of Select Importing Nations", fontsize=18, fontweight='bold')
ax.set_ylabel("Consumption (60Kg bags)", fontsize=12)

ax2 = ax.twinx()

bar2 = ax2.bar(
    x+width/2,
    price_cons["Price"],
    width = width,
    color = "orange"
)
ax.set_ylabel("Average Price Paid Per Bag (USD)", fontsize=12);

plt.xticks(x, price_cons["country"], fontsize = 7);

ax.legend([bar1, bar2], ["Imports", "Price"])
```
