```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [x for x in range(10)]
```

```python
fig, ax = plt.subplots()

ax.plot(data);
```

```python
plt.rc("axes.spines", top=False, right=False)
plt.rcParams["lines.linestyle"] = "--" 

fig, ax = plt.subplots()

ax.plot(data); 
```
