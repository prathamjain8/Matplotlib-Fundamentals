```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [x*10 for x in range(10)]
data
```

```python
labels = [f'{x}!' for x in range(10)]
labels
```

```python
plt.plot(labels, data);
plt.title("Title")
```

```python
fig = plt.figure();
ax = fig.add_subplot();

ax.plot(data);
ax.set_title("Title");
fig.suptitle("Overall Title")
```
