```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
coffee_production = pd.read_csv("D:/STUDIES/Data Visualization/Data/coffee/total-production.csv").T
coffee_production.columns = coffee_production.iloc[0]
coffee_production = coffee_production.drop("total_production")
```

```python
coffee_production.head()
```

```python
axes_settings = {
    "spines.top":False,
    "spines.right":False
}
plt.rc("axes", **axes_settings)
```

```python
brazil_vs_others = (coffee_production.assign(rest_of_world = coffee_production.drop("Brazil", axis=1).sum(axis=1))\
                                        .loc[:,["Brazil", "rest_of_world"]].astype({"Brazil":"float64"}))
brazil_vs_others.head()
```

```python
coffee_production["rest_of_world"] = coffee_production.drop(["Brazil", "Viet Nam", "Colombia", "Indonesia", "Ethiopia"]
                                                            , axis = 1).sum(axis=1)

coffee_production_top5_2018 = coffee_production.loc['2018',["Brazil", "Viet Nam", "Colombia", "Indonesia",
                                                            "Ethiopia", "rest_of_world"]]

pd.DataFrame(coffee_production_top5_2018)
```

# MeshGrid layout

```python
fig = plt.figure(figsize=(11,8.5))

gs = GridSpec(12, 6)

fig.suptitle("Brazil Exted its Dominance In Global Coffee Markets", fontsize = 20, fontweight='bold')

fig.text(.05, .683,
'''Brazil is the world's top coffee producer.





Since 1990, Brazil's share of global
production has risen from 20% to 29%,
even the worldwide market has 
expanded over 50%
'''         
, fontsize=12, fontweight = 'light')

ax1 = fig.add_subplot(gs[0:4, 2:4])
ax1.pie(
    brazil_vs_others.iloc[0].sort_values(ascending=False),
    startangle = 90,
    labels=["",""],
    pctdistance=.85,
    colors=["White", "#009639"]
    
)
center_circle = plt.Circle((0,0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.text(
    0,
    0,
    f"{round((brazil_vs_others.loc['1990', 'Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%",
    ha="center",
    va="center",
    fontsize = 28
)
ax1.set_title("Brazil % of global production 1990", fontsize=11)


ax2 = fig.add_subplot(gs[0:4, 4:])
ax2.pie(
    brazil_vs_others.iloc[-1].sort_values(ascending=False),
    startangle = 90,
    labels=["",""],
    pctdistance=.85,
    colors=["White", "#009639"]
    
)
center_circle = plt.Circle((0,0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.text(
    0,
    0,
    f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%",
    ha="center",
    va="center",
    fontsize = 28
)
ax2.set_title("Brazil % of global production 2018", fontsize=11)

ax3 = fig.add_subplot(gs[4:7, :])
ax3.stackplot(
    brazil_vs_others.index.astype("datetime64[ns]"),
    brazil_vs_others["Brazil"],
    brazil_vs_others["rest_of_world"].astype('float64'),
    labels=["Brazil", "World Total"],
    colors=["#009639","grey"]
)
ax3.set_title("Brazil Share of Market Grows as Global Output Surges")
ax3.set_ylabel("Production (60Kg Bags)", fontsize=12)
ax3.legend(loc="upper left")

ax4 = fig.add_subplot(gs[8:, 0:2])
ax4.barh(
    y = coffee_production_top5_2018.index[::-1],
    width=coffee_production_top5_2018.iloc[::-1],
    color=["grey","#009A44","#EF3340","#FFCD00","#C8102E","#009639"]
)
positions = list(range(6))
labels = ["BR","VN","CO","ID","ET","Rest Of World"][::-1]

ax4.set_title("Top Coffee Producing Nations 1990-2018")
ax4.set_xlabel("Production (Millions 60Kg Bags)", fontsize=10)
ax4.set_xticks(positions)
ax4.set_yticklabels(labels)

ax5 = fig.add_subplot(gs[8:, 2:])
ax5.pie(
    coffee_production_top5_2018[::-1],
    autopct="%.0f%%",
    labels=list(coffee_production_top5_2018.index[::-1]),
    startangle=90,
    colors=["grey","#009A44","#EF3340","#FFCD00","#C8102E","#009639"],
    explode=(0,0,0,0,0,.15)
)
ax5.set_title("Share of Global Coffee Production 2018")
```
